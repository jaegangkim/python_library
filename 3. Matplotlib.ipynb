```python
# data[data["사고유형_대분류"] == '차대차'][["발생지시도",'사상자수']]

temp=data.loc[data["사고유형_대분류"]=="차대차",['발생지시도','사상자수']]
```

```python
temp['발생지시도'].unique()
```

groupby

```python
y = result2['사상자수']       ## 시리즈면 result2.copy() 하면 됨
x = result2.index
plt.figure(figsize=(15,8))
plt.bar(x,y)
plt.title("차vs차 교통사고의 사상자 수")
plt.ylabel("사상자 수")
plt.xlabel("지역")
plt.show()
```

#### 실습2 : 교통사고가 가장 많이 발생하는 시간대를 알아보고 시각화
- 시간대 별로 카운트

```python
time = data["발생년월일시"]%100
# data['발생년월일시'].astype(str).str[-2:].astype(int)     똑같음
```

```python
time
```

```python
bins = [-1,2,5,8,11,14,17,20,23]
labels = ["0~2","3~5","6~8","9~11","12~14","15~17","18~20","21~23"]
result3 = pd.cut(time, bins=bins,labels =labels)
result3 = result3.value_counts().sort_index()
result3
```

```python
plt.figure(figsize=(5,5))
plt.pie(result3,labels=labels, autopct='%1.1f%%')
plt.show()
```

#### 실습4
- 주간 야간별로 발생한 사고 시각화
- 주간 경도, 주간 위도, 야간 경도, 야간 위도

```python
afternoon = data.loc[data['주야']=="주간",['경도','위도']]
night = data.loc[data['주야']=="야간",['경도','위도']]
```

```python
# 시각화 , scatter 사용
x = afternoon["경도"]
y = afternoon["위도"]
plt.scatter(x,y, alpha=0.1)  
plt.xlim(123,133)
plt.show()
```

```python
x = night["경도"]
y = night["위도"]
plt.scatter(x,y, alpha=0.1)  
plt.xlim(123,133)
plt.show()
```

```python
# 둘 다 한번에 표시
x1 = afternoon["경도"]
y1 = afternoon["위도"]
x2 = night["경도"]
y2 = night["위도"]
plt.scatter(x1,y1, alpha=0.1,c='r')
plt.scatter(x2,y2, alpha=0.1,c='b')
plt.xlim(123,133)
plt.show()
```
